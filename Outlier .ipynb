```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
music = pd.read_csv("Desktop\Project\spotify.csv")
music
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(music['popularity'])
plt.subplot(1,2,2)
sns.distplot(music['duration_ms'])
plt.show()
```

```python
sns.boxplot(music['duration_ms'])
```

```python
percentile25 = music['duration_ms'].quantile(0.25)
percentile75 = music['duration_ms'].quantile(0.75)
```

```python
iqr = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
music[music['duration_ms'] > upper_limit]
music[music['duration_ms'] < lower_limit]
```

```python
new_df = music[music['duration_ms'] < upper_limit]
new_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(music['duration_ms'])
plt.subplot(2,2,2)
sns.boxplot(music['duration_ms'])
plt.subplot(2,2,3)
sns.distplot(new_df['duration_ms'])
plt.subplot(2,2,4)
sns.boxplot(new_df['duration_ms'])
plt.show()
```

```python
new_df_cap = music.copy()
new_df_cap['duration_ms'] = np.where(new_df_cap['duration_ms'] > upper_limit, upper_limit,
    np.where(new_df_cap['duration_ms'] < lower_limit, lower_limit, new_df_cap['duration_ms']))
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(music['duration_ms'])
plt.subplot(2,2,2)
sns.boxplot(music['duration_ms'])
plt.subplot(2,2,3)
sns.distplot(new_df_cap['duration_ms'])
plt.subplot(2,2,4)
sns.boxplot(new_df_cap['duration_ms'])
plt.show()
```
